```python
import pandas as pd
```

```python
df = pd.read_csv('../Train_knight.csv')
```

```python
pd.set_option('display.max_columns', None)

pd.set_option('display.max_rows', None)

pd.set_option('display.width', 1000)

stats = df.describe().loc[['min', 'max', 'mean']]
print(stats)
```

```python
#Stims = 250
#Burst, Empowered, Friendship, Reactivity, Dexterity
target_feature = 'Stims'
other_features = ['Burst', 'Empowered', 'Friendship', 'Reactivity', 'Dexterity']

import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
cols_count = 5
rows_count = math.ceil(len(other_features) / cols_count)

# 3. Create the figure
fig, axes = plt.subplots(rows_count, cols_count, figsize=(20, rows_count * 4))
axes = axes.flatten() # Flatten to loop easily

for i, feature in enumerate(other_features):
    sns.scatterplot(ax=axes[i], data=df, x=feature, y=target_feature, 
                    hue='knight', palette={'Jedi': 'blue', 'Sith': 'red'}, alpha=0.6)
    axes[i].set_title(f'{target_feature} vs {feature}')
    axes[i].legend(title='Knight', loc='upper right')

# 4. Remove any empty subplots at the end
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
```

```python
# Midi-chlorien, Push
sns.scatterplot(data=df, x='Push', y='Midi-chlorien', hue='knight', palette={'Jedi': 'blue', 'Sith': 'red'}, alpha=0.6, legend='full')
plt.xlabel('Push')
plt.ylabel('Midi-chlorien')
plt.title('Midi-chlorien vs Push')
plt.show()
```
